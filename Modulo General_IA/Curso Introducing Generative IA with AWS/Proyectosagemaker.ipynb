# 🧠 Proyecto PartyRock: **Su Merced AgroAsiste**

---

## 💡 Nombre Tentativo de la App

**Su Merced AgroAsiste**
*Un asistente rural en lenguaje cercano para apoyar la comercialización y las finanzas agrícolas de los pequeños productores colombianos.*

---

## 🧱 1. Estructura general del proyecto en PartyRock

### 🎯 **Propósito de la app**

Brindar **orientación financiera y comercial** a pequeños productores agrícolas en Colombia de forma **cercana, accesible y útil**, para que puedan:

* Entender sus **costos de producción**.
* Identificar alternativas de **comercialización justa**.
* Conocer **precios estimados de mercado**.
* Recibir sugerencias para **agregar valor** a sus productos.
* Acceder a herramientas o ideas de **autogestión económica rural**.

---

### 👥 **Audiencia**

Pequeños y medianos productores rurales colombianos, especialmente en zonas con:

* Bajo acceso a asistencia técnica o financiera.
* Bajo nivel de escolaridad.
* Acceso digital limitado (uso exclusivo o frecuente de **teléfonos celulares**).

---

## 📲 2. Diseño de la aplicación en PartyRock

### 🧾 **Entradas del usuario (Inputs)**

| Entrada                                           | Tipo           | Ejemplo                      |
| ------------------------------------------------- | -------------- | ---------------------------- |
| ¿Qué está cultivando su merced?                   | Texto libre    | Yuca, café, frijol           |
| ¿En qué etapa va el cultivo?                      | Lista o texto  | Sembrado / Listo pa' vender  |
| ¿Cuánto ha invertido más o menos?                 | Texto o número | 800.000 pesos                |
| ¿Cuánto espera cosechar (en kilos o arrobas)?     | Número         | 12 arrobas                   |
| ¿Dónde se encuentra su finca? (vereda, municipio) | Texto          | Vereda Portachuelo, Rionegro |

---

### 💬 **Salidas generadas (Outputs)**

* Estimación del **costo por unidad de producto**.
* Estimación de **ganancia o pérdida probable**.
* Recomendaciones de **lugares cercanos para vender**, ideas de empaque, ferias, plazas o asociaciones.
* Ideas para **agregar valor** (ej. transformar yuca en almidón, café en bolsitas tostadas).
* Sugerencias sobre **precios de mercado aproximados**, según el producto y la región.
* Recomendaciones sobre **cómo llevar un cuaderno de gastos sencillo**.

---

### 🗣️ **Tono y accesibilidad**

* Usar **lenguaje campesino, cercano, sin tecnicismos**.
* Palabras como: *su merced, pa’ que no pierda la platica, con eso no se lo pagan a huevo, puede ir a tal parte a vender, eso lo paga bueno*.
* Interfaz simple, accesible desde **teléfono móvil**.
* Considerar agregar íconos visuales o separar salidas en bloques claros.

> *Crea una aplicación en tono campesino colombiano que ayude a pequeños productores rurales a organizar sus finanzas del cultivo y saber cómo y dónde vender su producción. ## 🌱 5. Ideas de prueba y mejora

### Casos realistas:

* Usuario 1: Cultiva papa en Boyacá, sembrado hace un mes.
* Usuario 2: Tiene café listo pa’ vender, no sabe cuánto pedir por arroba.
* Usuario 3: Invierte mucho pero siempre le toca mal vender, quiere opciones.

### Casos extremos:

* Usuario pone: “no sé qué hacer”, “no sé cuánto cosecho”.
* Usuario escribe mal el cultivo: “yucca”, “pappa”.
